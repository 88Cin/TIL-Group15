```python
# 定义文件夹路径
ADF_folder_path = 'data/ADF/'
HH_file_path = 'data/HH/multi_size.csv'  # 固定的HH文件路径

# 使用os.listdir()获取ADF文件夹内的所有文件
ADF_files = [f for f in os.listdir(ADF_folder_path) if f.endswith('.csv')]

df2 = pd.read_csv(HH_file_path)  # 事先读取固定的HH文件

for ADF_file in ADF_files:
    file_path1 = os.path.join(ADF_folder_path, ADF_file)
    df1 = pd.read_csv(file_path1)

    df1['x_lead_follow_diff'] = df1['x_lead'] - df1['x_follow']
    mean_x_lead_follow_diff = df1.groupby('ID')['x_lead_follow_diff'].mean()

    # 合并df1和df2，使用ID列进行匹配
    df_combined = df1.merge(df2, on='ID')

    # 计算half_size_sum，使用size_lead和size_follow
    half_size_sum = (df_combined['size_lead'] + df_combined['size_follow']) * 0.5
    s = mean_x_lead_follow_diff - half_size_sum

    car_0_to_1000 = df_combined
    mean_a_follow = df_combined.groupby('ID')['v_follow'].mean()

    sns.set(style="whitegrid")
    g = sns.jointplot(x=s, y=mean_a_follow, data=car_0_to_1000, kind="scatter", height=7)

    g.fig.suptitle(f'Scatter Plot for Cars Gap-Velocity({ADF_file.split(".")[0]})', y=1.02)
    g.set_axis_labels("Gap(m)", "Mean velocity of followed cars(m/s)")

    plt.show()
```
